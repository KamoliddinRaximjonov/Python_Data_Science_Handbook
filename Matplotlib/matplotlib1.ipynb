```python
import numpy as np
import matplotlib.pyplot as plt

# vaqt oqi (sekund)
t = np.linspace(0, 1, 1000)

# yurak impulsi (P-QRS-T tolqinlariga oxshash)
signal = (
    0.1 * np.sin(2 * np.pi * 5 * t) +             # P-tolqin
    np.exp(-((t - 0.3) / 0.02)**2) * -0.15 +      # Q
    np.exp(-((t - 0.32) / 0.01)**2) * 1.2 +       # R (asosiy impuls)
    np.exp(-((t - 0.35) / 0.02)**2) * -0.3 +      # S
    np.exp(-((t - 0.6) / 0.05)**2) * 0.3          # T-tolqin
)

plt.figure(figsize=(4, 4))
plt.plot(t, signal)
plt.xlabel("Vaqt (sekund)")
plt.ylabel("Amplituda")
plt.title("Yurak urishi (EKG) signali")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parametrlar
fs = 1000          # sampling rate (Hz)
duration = 5       # signal davomiyligi (sekund)
t = np.linspace(0, duration, fs * duration)

heart_rate = 75    # BPM
period = 60 / heart_rate

signal = np.zeros_like(t)

# Har bir yurak urishini qoshamiz
for beat in np.arange(0, duration, period):
    signal += (
        0.1 * np.exp(-((t - (beat + 0.1)) / 0.03)**2) +   # P
        -0.15 * np.exp(-((t - (beat + 0.2)) / 0.01)**2) + # Q
        1.2 * np.exp(-((t - (beat + 0.22)) / 0.008)**2) + # R
        -0.3 * np.exp(-((t - (beat + 0.24)) / 0.01)**2) + # S
        0.3 * np.exp(-((t - (beat + 0.4)) / 0.04)**2)     # T
    )

plt.figure(figsize=(12, 4))
plt.plot(t, signal)
plt.title("Ketma-ket yurak urishlari (EKG)")
plt.xlabel("Vaqt (sekund)")
plt.ylabel("Amplituda")
plt.grid(True)
plt.show()
```

```python
# Shovqin qoshamiz (real signalga yaqinlashtirish)
```

```python
noise = 0.05 * np.random.randn(len(signal))
noisy_signal = signal + noise

plt.figure(figsize=(12, 4))
plt.plot(t, noisy_signal)
plt.title("Shovqinli yurak signali (realga yaqin)")
plt.xlabel("Vaqt (sekund)")
plt.ylabel("Amplituda")
plt.grid(True)
plt.show()
```

```python
#Yurak urishini aniqlash (R-peak) va BPM hisoblash
```

```python
threshold = 0.8
peaks = np.where(noisy_signal > threshold)[0]

# yaqin nuqtalarni bitta peak deb olish
filtered_peaks = [peaks[0]]
for p in peaks:
    if p - filtered_peaks[-1] > fs * 0.3:
        filtered_peaks.append(p)

peak_times = np.array(filtered_peaks) / fs
bpm = 60 / np.mean(np.diff(peak_times))

print(f"Hisoblangan BPM: {bpm:.2f}")
```

```python
t = np.linspace(0, 1, 1000) # bir sekundlik vaqt ms da



# Suniy yurak impulsi formulasi
signal = (
    0.1 * np.sin(2 * np.pi * 5 * t) +          # P-tolqin
    np.exp(-((t - 0.3)/0.02)**2) * -0.15 +     # Q
    np.exp(-((t - 0.32)/0.01)**2) * 1.2 +      # R (asosiy urish)
    np.exp(-((t - 0.35)/0.02)**2) * -0.3 +     # S
    np.exp(-((t - 0.6)/0.05)**2) * 0.3         # T-tolqin
)


# exp(-((t-a)/b)^2)  Gauss tolqini, R-peak kabi choqqilar uchun
# sin()  P-tolqin uchun


plt.figure(figsize=(2, 4))  # olcham
plt.plot(t, signal)           # signalni chizish
plt.title("Yurak urishi (EKG) signali")
plt.xlabel("Vaqt (sekund)")
plt.ylabel("Amplituda")
plt.grid(True)                # grid qoshish
plt.show()
```
