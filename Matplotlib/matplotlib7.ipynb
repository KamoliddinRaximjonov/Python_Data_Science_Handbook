Customizing Plot Legends

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use("classic")
```

```python
x = np.linspace(0, 10, 1000) # 0 dan 10 gancaha bolagan sonlarni 1000 ta oraliqqa ajratadi

fig, ax = plt.subplots()
ax.plot(x, np.sin(x), "-b", label = "Sine")
ax.plot(x, np.cos(x), "--r", label = "Cosine")
ax.axis("equal")
leg = ax.legend()
plt.show()  # Standart (default) grafik legendasi
```

```python
ax.legend(loc = "upper left", frameon = False) # Moslashtirilgan grafik legendasi
fig
```

```python
ax.legend(frameon = False, loc = "lower center", ncol = 2)  # Ikki ustunli grafik legendasi
fig
```

```python
ax.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)
fig
# Fancybox uslubidagi grafik legendasi
```

```python
"""                             Choosing Elements for the Legend                            """
```

```python
y = np.sin(x[:, np.newaxis] + np.pi * np.arange(0, 2, 0.5))
lines = plt.plot(x, y)

# lines  plt.Line2D obyektlaridan iborat royxat

plt.legend(lines[:2], ["first", "second"])
plt.show() # Legenda elementlarini moslashtirish
```

```python
plt.plot(x, y[:, 0], label='first')
plt.plot(x, y[:, 1], label='second')
plt.plot(x, y[:, 2:])
plt.legend(framealpha=1, frameon=True)
plt.show()
```

```python
import pandas as pd
cities = pd.read_csv('california_cities.csv')

# Bizga kerakli ma'lumotlarni ajratib olish
lat, lon = cities['latd'], cities['longd']
population, area = cities['population_total'], cities['area_total_km2']

# Nuqtalarni chizish (olcham va rang bor, lekin label yoq)
plt.scatter(lon, lat, label=None,
            c=np.log10(population), cmap='viridis',
            s=area, linewidth=0, alpha=0.5)

plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3, 7)

# Bu yerda legendani yaratamiz:
# kerakli olcham va label bilan bosh nuqtalar chizamiz
for area in [100, 300, 500]:
    plt.scatter([], [], c='k', alpha=0.3, s=area,
                label=str(area) + ' km$^2$')

plt.legend(scatterpoints=1, frameon=False,
           labelspacing=1, title='City Area')

plt.title('California Cities: Area and Population')
plt.show()
```

```python
fig, ax = plt.subplots()
lines = []
styles = ['-', '--', '-.', ':']
x = np.linspace(0, 10, 1000)

for i in range(4):
    lines += ax.plot(x, np.sin(x - i * np.pi / 2),
                     styles[i], color='black')

ax.axis('equal')

# Birinchi legendaning chiziqlari va yorliqlari
ax.legend(lines[:2], ['line A', 'line B'],
          loc='upper right', frameon=False)

# Ikkinchi legendani yaratish va uni qolda qoshish
from matplotlib.legend import Legend
leg = Legend(ax, lines[2:], ['line C', 'line D'],
             loc='lower right', frameon=False)

ax.add_artist(leg)
plt.show()
```

```python
df = pd.read_csv("summer2016.csv")
df.head()
```

```python
df.isna().sum()
```

```python
df_fil = df.dropna(subset= ["ID", "Name", "Sex", "Height", "Weight"])
df_fil.head()
```

```python
gr = df_fil.groupby(["NOC"])["Sport"].count()
plt.figure(figsize= (12, 6))
plt.hist(x= gr)
plt.show()
```

```python
gr = df_fil.groupby(["NOC"])["Sport"].count()
plt.figure(figsize= (20, 10))
plt.plot(gr, marker = "o")
plt.xticks(rotation = 90)
plt.show()
```

```python
df_pt = df_fil.pivot_table(columns= "Sex", index= "NOC", values= "Age", aggfunc= "mean",
                           fill_value= 0)
df_pt.head()
```

```python
fig, ax = plt.subplots(2, figsize= (20, 10))

ax[0].plot(df_pt.index, df_pt["M"], marker = "o", color = "b", label = "Male")
ax[0].tick_params(axis='x', rotation=90)
ax[0].set_title('Average Age: Male')
ax[0].set_ylabel('Age')

ax[1].plot(df_pt.index, df_pt["F"], marker = "o",color = "red", label = "Female")
ax[1].tick_params(axis='x', rotation=90)
ax[1].set_title('Average Age: Female')
ax[1].set_ylabel('Age')

plt.legend(title = "Sex")
plt.xticks(rotation = 90)
plt.xlabel("Country (NOC)")
plt.show()
```
