Density and Contour Plots

```python
"""                           Visualizing a Three-Dimensional Function                       """
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.style.use("ggplot")
```

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)

Z = f(X, Y)
```

```python
plt.contour(X, Y, Z, colors = "black")
plt.show()
```

```python
# rangli kontur chizish 
plt.contour(X, Y, Z, 20, cmap = "RdGy")
plt.show()
```

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()
plt.show()
```

```python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='RdGy', alpha=0.5)
plt.colorbar()
plt.show()
```

```python
def func(x1, y1):
    return np.cos(x1)  ** 23 + np.sin(x1) * np.cos(y1) - (np.cos(y1) ** 2 + np.sin(x1) ** 2)
```

```python
x1 = np.linspace(0, 8, 80)
y1 = np.linspace(0, 8, 60)

X1, Y1 = np.meshgrid(x1, y1)

Z1 = func(X1, Y1)

```

```python
plt.contour(X1, Y1, Z1, color = "black")
plt.show()
```

```python
plt.contour(X1, Y1, Z1, 20, cmap = "RdGy")
plt.show()
```

```python
plt.contourf(X1, Y1, Z1, 20, cmap = "RdGy")
plt.colorbar()
plt.show()
```

```python
plt.imshow(Z1, extent=[0, 5, 0, 5], origin='lower',
           cmap='RdGy')
plt.colorbar()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
contours = plt.contour(X1, Y1, Z1, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z1, extent=[0, 5, 0, 5], origin='lower',
           cmap='RdGy', alpha=0.5)
plt.colorbar()
plt.show()
```
