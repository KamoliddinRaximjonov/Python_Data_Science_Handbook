```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
"""                               plt.axes: Subplots by Hand                                 """
```

```python
ax1 = plt.axes()
ax2 = plt.axes([0.65, 0.65, 0.2, 0.2])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4], xticklabels = [], ylim = (-1.2, 1.2))
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4], ylim = (-1.2, 1.2))
x = np.linspace(0, 10)
ax1.plot(np.sin(x))
ax2.plot(np.cos(x))
plt.show()
```

```python
"""                           plt.subplot: Simple Grids of Subplots                          """
```

```python
for i in range(1, 7):
    plt.subplot(2, 3, i)
    plt.text(0.5, 0.5, str((2, 3, i)),
             fontsize = 18, ha ="center" )
```

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace= 0.4)

for i in range(1, 7):
    ax= fig.add_subplot(2, 3, i)
    ax.text(0.5, 0.5, str((2, 3, i)),
            fontsize = 18, ha = "center")
```

```python
"""                        plt.subplots: The Whole Grid in One Go                             """
```

```python
fig, ax = plt.subplots(2, 3, sharex='col', sharey='row')
```

```python
# oqlar ikki olchamli massivda joylashgan: [qator, ustun]
for i in range(2):
    for j in range(3):
        ax[i, j].text(0.5, 0.5, str((i, j)),
                      fontsize=18, ha='center')

fig
```

```python
"""                         plt.GridSpec: More Complicated Arrangements                      """
```

```python
grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.3)
```

```python
plt.subplot(grid[0, 0])
plt.subplot(grid[0, 1:])
plt.subplot(grid[1, :2])
plt.subplot(grid[1, 2])
```

```python
# Normal taqsimlangan malumotlarni yaratish
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 3000).T

# GridSpec yordamida oqlarni sozlash
fig = plt.figure(figsize=(6, 6))
grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)

main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

# Asosiy oqda scatter plot
main_ax.plot(x, y, 'ok', markersize=3, alpha=0.2)

# Yon oqlarda gistogrammalar
x_hist.hist(x, 40, histtype='stepfilled',
            orientation='vertical', color='gray')
x_hist.invert_yaxis()

y_hist.hist(y, 40, histtype='stepfilled',
            orientation='horizontal', color='gray')
y_hist.invert_xaxis()
```

```python
df = pd.read_csv("summer2016.csv")
df.head()
```

```python
df_fil = df.dropna()
df_fil.head()
```

```python
dfT = df_fil.pivot_table(index= "NOC", columns= "Medal", values= "Sex", aggfunc= "size", fill_value= 0)
dfT.head()
```

```python
fig, ax = plt.subplots(figsize = (20, 10))
ax.plot(dfT.index, dfT["Gold"], color = "gold", marker = "o", label = "Gold")
ax.plot(dfT.index, dfT["Silver"], color = "silver", marker = "o", label = "Silver")
ax.plot(dfT.index, dfT["Bronze"], color = "brown", marker = "o", label = "Bronze")
plt.legend(title = "Medals")
plt.xlabel("NOCs")
plt.ylabel("Count of medals")
plt.xticks(rotation = 90)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (20, 10))
ax.bar(dfT.index, dfT["Gold"], color = "gold", label = "Gold")
ax.bar(dfT.index, dfT["Silver"], color = "silver", label = "Silver")
ax.bar(dfT.index, dfT["Bronze"], color = "brown", label = "Bronze")
plt.legend(title = "Medals")
plt.xlabel("NOCs")
plt.ylabel("Count of medals")
plt.xticks(rotation = 90)
plt.show()
```

```python
fig, ax = plt.subplots(3, 1, figsize = (20, 10))

ax[0].plot(dfT.index, dfT["Gold"], color = "gold", marker = "o", label = "Gold")
ax[0].tick_params(rotation=90)


ax[1].plot(dfT.index, dfT["Silver"], color = "silver", marker = "o", label = "Silver")
ax[1].tick_params(rotation=90)

ax[2].plot(dfT.index, dfT["Bronze"], color = "brown", marker = "o", label = "Bronze")
ax[2].tick_params(rotation=90)


plt.xlabel("NOCs")
plt.ylabel("Count of medals")
plt.xticks(rotation = 90)
plt.show()
```

```python
fig, ax = plt.subplots(3, 1, figsize = (20, 20))

ax[0].bar(dfT.index, dfT["Gold"], color = "gold", label = "Gold")
ax[0].set_title("GOLD")
ax[0].tick_params(rotation=90)
ax[0].set_ylabel('Count of medals')


ax[1].bar(dfT.index, dfT["Silver"], color = "silver", label = "Silver")
ax[1].set_title("SILVER")
ax[1].tick_params(rotation=90)
ax[1].set_ylabel('Count of medals')

ax[2].bar(dfT.index, dfT["Bronze"], color = "brown", label = "Bronze")
ax[2].set_title("BRONZE")
ax[2].tick_params(rotation=90)
ax[2].set_ylabel('Count of medals')


plt.xlabel("NOCs")
plt.ylabel("Count of medals")
plt.xticks(rotation = 90)
plt.show()
```
