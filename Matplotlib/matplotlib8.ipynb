```python
from matplotlib.colors import LinearSegmentedColormap

def grayscale_cmap(cmap):
    
    """Berilgan rang xaritasining kulrang variantini qaytaradi"""
    cmap = plt.cm.get_cmap(cmap)
    colors = cmap(np.arange(cmap.N))

    # RGBA  inson kozi sezadigan kulrang yoruglik
    RGB_weight = [0.299, 0.587, 0.114]
    luminance = np.sqrt(np.dot(colors[:, :3] ** 2, RGB_weight))
    colors[:, :3] = luminance[:, np.newaxis]

    return LinearSegmentedColormap.from_list(
        cmap.name + "_gray", colors, cmap.N
    )

def view_colormap(cmap):

    """Rang xaritasi va uning kulrang ekvivalentini chizadi"""
    cmap = plt.cm.get_cmap(cmap)
    colors = cmap(np.arange(cmap.N))
    cmap = grayscale_cmap(cmap)
    grayscale = cmap(np.arange(cmap.N))

    fig, ax = plt.subplots(2, figsize=(6, 2),
                           subplot_kw=dict(xticks=[], yticks=[]))
    ax[0].imshow([colors], extent=[0, 10, 0, 1])
    ax[1].imshow([grayscale], extent=[0, 10, 0, 1])
```

```python
view_colormap('jet')
```

Example: Handwritten Digits

```python
from sklearn.datasets import load_digits
digits = load_digits(n_class=6)  # 0 dan 5 gacha bolgan raqamlar
fig, ax = plt.subplots(8, 8, figsize=(6, 6))
for i, axi in enumerate(ax.flat):
    axi.imshow(digits.images[i], cmap='binary')
    axi.set(xticks=[], yticks=[])
```

```python
from sklearn.manifold import Isomap
iso = Isomap(n_components=2)
projection = iso.fit_transform(digits.data)
```

```python
plt.scatter(projection[:, 0], projection[:, 1], lw=0.1,
            c=digits.target, cmap=plt.cm.get_cmap('cubehelix', 6))
plt.colorbar(ticks=range(6), label='digit value')
plt.clim(-0.5, 5.5)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Misol DataFrame
data = {
    "NOC": ["USA", "USA", "FRA", "FRA", "CHN", "CHN", "CHN"],
    "Height": [180, 175, 172, 168, 185, 178, 180]
}
df_filr = pd.DataFrame(data)

# Guruhlash va sanash
df_gb1 = df_filr.groupby("NOC")["Height"].mean()

# Rank berish (katta count birinchi)
df_gb1_rank = df_gb1.rank(ascending=False)

# Rang berish uchun scatter plot (NOC indexlarini x, rank-ni c sifatida)
plt.figure(figsize=(6,4))
plt.scatter(df_gb1.index, df_gb1_rank, c=df_gb1_rank, cmap='viridis', s=300)
plt.colorbar(label='Rank')
plt.xlabel('NOC')
plt.ylabel('Rank')
plt.title('NOC boyicha Height count rank va colorbar')
plt.show()
```
