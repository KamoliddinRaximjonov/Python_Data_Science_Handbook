```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x= np.linspace(0, 10, 1000)

plt.figure() # garfik uvhun figura yasash

# ikki paneldan iborat garfikning binichisini yaratish va joriya o'qni tanlash 

plt.subplot(2, 1, 1) #(qatorlar, ustunlar, panel raqami)
plt.plot(x, np.sin(x))

# # ikkinchi panelni yaratish va joriy o‘qni tanlash
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))

plt.show()

```

```python
# Avval subgrafiklar panjarasini yaratamiz
# ax — ikkita Axes obyektidan iborat massiv

fig, ax = plt.subplots(2)

# Tegishli obyektning plot() metodini chaqiramiz
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
plt.show()

```

```python
def func(x_max, k):
    x = np.linspace(0, x_max, 1000)
    y = k * x

    if k > 0:
        plt.subplot(2, 1, 1)
        plt.plot(x, y)
        plt.title("k > 0 (musbat chiziq)")
        
    else:
        plt.subplot(2, 1, 2)
        plt.plot(x, y)
        plt.title("k < 0 (manfiy chiziq)")


func(4, -5)
func(4, 5)
plt.tight_layout()
plt.show()
```

```python
def func2(k, x_max2, b):
    x = np.linspace(0, x_max2, 1000)
    y = k * x + b

    if k > 0:
        plt.subplot(2, 1, 1)
        plt.plot(x, y)
        plt.title("k > 0 (musbat chiziq)")
        
    else:
        plt.subplot(2, 1, 2)
        plt.plot(x, y)
        plt.title("k < 0 (manfiy chiziq)")
        
func2(2, 4, 6 )
func2(-2, 4, -6 )

plt.tight_layout()
plt.show()
```

```python
def func3(k, x_max2):
    
    plt.figure(figsize=(12, 15))
    x = np.linspace(0, x_max2, 100)
    y = k / x

    if k > 0:
        plt.subplot(2, 1, 1)
        plt.plot(x, y)
        plt.title("k > 0 (musbat chiziq)")
        
    else:
        plt.subplot(2, 1, 2)
        plt.plot(x, y)
        plt.title("k < 0 (manfiy chiziq)")
        
func3(2, 4)
func3(-2, 4)

plt.tight_layout()
plt.show()
```

```python
def func4(x_max2):
    
    plt.figure(figsize=(12, 15))

    x = np.linspace(0, x_max2, 10000)
    y = np.sqrt(x)

    plt.subplot(2, 1, 1)
    plt.plot(x, y)

func4(4)

plt.tight_layout()
plt.show()
```

```python
def func5(x_max2):
    
    plt.figure(figsize=(12, 15))

    x = np.linspace(0, x_max2, 10000)
    y = np.pow(x, 1/ 3)

    plt.subplot(2, 1, 1)
    plt.plot(x, y)

func5(4)

plt.tight_layout()
plt.show()
```

```python
def funcU(a, k, b):
    x = np.linspace(-5, 5, 400)

    fig, ax = plt.subplots(12, 1, figsize=(6, 18))

    ax[0].plot(x, np.sin(x)); ax[0].set_title("sin(x)")
    ax[1].plot(x, np.cos(x)); ax[1].set_title("cos(x)")
    ax[2].plot(x, np.tan(x)); ax[2].set_title("tan(x)")
    ax[3].plot(x, np.cos(x)/np.sin(x)); ax[3].set_title("cot(x)")
    ax[4].plot(x, np.cbrt(x)); ax[4].set_title("x^(1/3)")

    if a > 1:
        ax[5].plot(x, a**x)
        ax[5].set_title("a^x (a > 1)")
    else:
        ax[6].plot(x, a**x)
        ax[6].set_title("a^x (0 < a < 1 or a < 0)")

    if a > 0:
        ax[7].plot(x, a * x**2)
        ax[7].set_title("a·x² (a > 0)")
    else:
        ax[8].plot(x, a * x**2)
        ax[8].set_title("a·x² (a < 0)")

    if k < 0:
        ax[9].plot(x, k * x)
        ax[9].set_title("k·x (k < 0)")
    else:
        ax[10].plot(x, k * x)
        ax[10].set_title("k·x (k > 0)")

    if k < 0:
        ax[11].plot(x, k * x + b)
        ax[11].set_title("k·x + b (k < 0)")
    else:
        ax[11].plot(x, k * x + b)
        ax[11].set_title("k·x + b (k > 0)")

    plt.tight_layout()

funcU(2, 3, 6)
funcU(-2, -3, 6)

plt.show()
```

```python
"""                          Adjusting the Plot: Line Colors and Styles                     """
```

```python
plt.plot(x, np.sin(x - 0), color='blue')        # rang nomi bilan
plt.plot(x, np.sin(x - 1), color='g')           # qisqa rang kodi (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # kulrang (0 va 1 oralig‘ida)
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex kodi (00–FF, RRGGBB)
plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3)) # RGB tuple (0–1 oralig‘ida)
plt.plot(x, np.sin(x - 5), color='chartreuse')  # HTML rang nomlari

plt.show()
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')

# Qisqa yozuvlar:
plt.plot(x, x + 4, linestyle='-')   # uzluksiz (solid)
plt.plot(x, x + 5, linestyle='--')  # uzilgan (dashed)
plt.plot(x, x + 6, linestyle='-.')  # nuqta-chiziq (dashdot)
plt.plot(x, x + 7, linestyle=':')   # nuqtali (dotted)
```

```python
plt.plot(x, x + 0, '-g')   # yashil, uzluksiz chiziq
plt.plot(x, x + 1, '--c')  # ko‘k-yashil (cyan), uzilgan chiziq
plt.plot(x, x + 2, '-.k')  # qora, nuqta-chiziq
plt.plot(x, x + 3, ':r')   # qizil, nuqtali chiziq
```

Matplotlib Gotchas
While most plt functions translate directly to ax methods (such as plt.plot() →
ax.plot(), plt.legend() → ax.legend(), etc.), this is not the case for all com‐
mands. In particular, functions to set limits, labels, and titles are slightly modified.
For transitioning between MATLAB-style functions and object-oriented methods,
make the following changes:


```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel='x', ylabel='sin(x)',
       title='A Simple Plot')
```
